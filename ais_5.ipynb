```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
```

```python
x = np.linspace(0, 10, 50)
y = x
```

```python
plt.title("Линейная зависимость y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()  #сетка
plt.plot(x, y)
```

```python
plt.title("Линейная зависимость y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()  #сетка
plt.plot(x, y, "r--")
```

```python
x = np.linspace(0, 10, 50)
y1 = x
y2 = [i ** 2 for i in x]
plt.title("Зависимости: y1 = x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()
plt.plot(x, y1, x, y2)
```

```python
x = np.linspace(0, 10, 50)
y1 = x
y2 = [i ** 2 for i in x]
plt.figure(figsize=(9, 9))
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("Зависимости: y1 = x, y2 = x^2")
plt.ylabel("y1", fontsize=14)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.xlabel("x", fontsize=14)
plt.xlabel("y2", fontsize=14)
plt.grid(True)
```
